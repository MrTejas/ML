```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('advertisement.csv')

col_names = ['gender', 'education','occupation','most bought item', 'children'] 
for col in col_names:
    plt.figure(figsize=(6, 3))
    sns.countplot(x=col, data=dataset)
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)
    plt.show()

# Countplot of boolean features
boolean_columns = ['married']  # Replace with actual column names
for col in boolean_columns:
    plt.figure(figsize=(6, 3))
    sns.countplot(x=col, data=dataset)
    plt.title(f'Countplot of {col}')
    plt.xticks([0, 1], ['False', 'True'])
    plt.show()
```

```python
dataset = pd.read_csv('advertisement.csv')

# 1. Scatter Plot of income, purchase_amount, and age
plt.figure(figsize=(4,3))
sns.scatterplot(x='income', y='purchase_amount', hue='age', data=dataset)
plt.title('Scatter Plot of Income, purchase_amount, and Age')
plt.xlabel('Income')
plt.ylabel('purchase_amount')
plt.legend(title='Age')
plt.show()

# 2. Histogram of age
plt.figure(figsize=(4,3))
sns.histplot(dataset['age'], bins=20, kde=True)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 3. Bar graph of average income and purchase_amount for each occupation label
occupation_stats = dataset.groupby('occupation')[['income', 'purchase_amount']].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='occupation', y='income', data=occupation_stats, color='b', label='Income')
sns.barplot(x='occupation', y='purchase_amount', data=occupation_stats, color='g', label='Purchase_Amount',
            saturation=0.5)  
# not working - debug
plt.title('Average Income and Purchase_Amount by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout() 
plt.show()


# 4. Bar graph of average income and purchase_amount for each education label
education_stats = dataset.groupby('education')[['income', 'purchase_amount']].mean().reset_index()
plt.figure(figsize=(5,3))
sns.barplot(x='education', y='income', data=education_stats, color='b', label='Income')
sns.barplot(x='education', y='purchase_amount', data=education_stats, color='g', label='purchase_amount')
plt.title('Average Income and purchase_amount by Education')
plt.xlabel('Education')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.legend()
plt.show()

```
